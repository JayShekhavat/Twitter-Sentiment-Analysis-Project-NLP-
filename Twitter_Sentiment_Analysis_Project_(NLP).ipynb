## Stage 1 ->data convert to lower case

```python
text = text.tolist()
```

```python
len(text)
```

```python
lower_case_stage_1 = []
def lower_case(text):
  for sent in text:
    lower_case_stage_1.append(str.lower(sent))
```

```python
print(type(lower_case_stage_1))
```

```python
len(lower_case_stage_1)
```

```python
lower_case(text)
```

```python
lower_case_stage_1
```

##stage 2 -> Tokenization

```python
from nltk.tokenize import sent_tokenize,word_tokenize
nltk.download('punkt')
```

```python
token_stage_2 = []
for sent in lower_case_stage_1:
    sent = word_tokenize(sent)
    token_stage_2.append(sent)
```

```python
token_stage_2
```

##Remove Special characters from dataset

```python
len(token_stage_2)
```

```python
import re
```

```python
removed_spl_char = []
for words in token_stage_2:
  clean = []
  for word in words:
    res = re.sub(r'[^\w\s]', "",word)
    if res !="":
      clean.append(res)
  removed_spl_char.append(clean)

```

```python
removed_spl_char
```

##Remove Stopwords from dataset

```python
removed_stop_words = []
for words in removed_spl_char:
  w = []
  for  word in words:
    if not word in stopwords.words('english'):
      w.append(word)
  removed_stop_words.append(w)


```

```python
removed_stop_words
```

##Stemming the words of sentences of dataset.

```python
stemming = PorterStemmer()
```

```python
stemmed_text = []
for sent in removed_stop_words:
  w = []
  for word in sent:
    w.append(stemming.stem(word))
  stemmed_text.append(w)
```

```python
stemmed_text
```

##Lemmetized the words of sentences of dataset.

```python
nltk.download('wordnet')
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
lemmetized_text = []
for sent in stemmed_text:
  w = []
  for word in sent:
    w.append(lemmatizer.lemmatize(word, pos = 'a'))
  lemmetized_text.append(" ".join(w))
```

```python
df['text'] = lemmetized_text
```
